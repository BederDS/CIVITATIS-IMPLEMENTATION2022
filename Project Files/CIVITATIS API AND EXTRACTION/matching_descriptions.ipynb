### Matching descriptions from CIVITATIS activities

In this script I take some activities from CIVITATIS that have mistakes in the description columns and correct them using the infomartion from the CIVITATIS API.

```python
# Install and import packages
import pandas as pd
import numpy as np
```

Load the table with mistakes and clean it

```python
# Sample taken from the civitatis_activities table
df_err=pd.read_excel("C:/Users/beder/Desktop/Data 2022/Projects/CIVITATIS-IMPLEMENTATION2022/Project Files/inputs/20221229_civitatis_erros.xlsx")
df_err.shape
```

```python
df_err=df_err[['civitatis_activity_id', 'civitatis_id', 'destination_civitatis_id',
       'city_id', 'place_id']]
```

```python
df_err
```

Load the table with information in english

```python
df_engl=pd.read_csv('C:/Users/beder/Desktop/Data 2022/Projects/CIVITATIS-IMPLEMENTATION2022/Outputs/Civitatis Exctraction/all_activities30_12_2022.csv')
df_engl.columns
```

```python
df_engl=df_engl[['id','title','description']]
df_engl.head(20)
```

```python
#df_engl.duplicated('id')
ids=df_engl['id']
df_engl[ids.isin(ids[ids.duplicated()])].sort_values("id")
```

```python
df_engl=df_engl.drop_duplicates()
df_engl.shape
```

```python
# Merge both tables and keep only ids of df_err
df_err_engl=df_err.merge(df_engl, how='left',left_on='civitatis_id',right_on='id')
df_err_engl.shape
```

```python
df_err_engl=df_err_engl.rename(columns={'id':'id_english','title':'name_en','description':'description_en'})
df_err_engl
```

Load activities in spanish

```python
df_es=pd.read_csv('C:/Users/beder/Desktop/Data 2022/Projects/CIVITATIS-IMPLEMENTATION2022/Outputs/Civitatis Exctraction/all_activities30_12_2022_es.csv')
df_es.columns
```

```python
df_es=df_es[['id','title','description']]
df_es=df_es.drop_duplicates()
df_es.shape
```

```python
df_err_engl_es=df_err_engl.merge(df_es, how='left',left_on='civitatis_id',right_on='id')
df_err_engl_es.shape
```

```python
df_err_engl_es
```

```python
df_err_engl_es=df_err_engl_es.rename(columns={'id':'id_espa','title':'name_es','description':'description_es'})
df_err_engl_es
```

Clean strange characters in the description

```python
# Replace strange characters in description_es and description_en

char=['<strong>','<p>','</p>','</strong>','<b>','</b>','</em>','&nbsp','&amp','<em>',';']

for i in char:
    df_err_engl_es['description_es'] = df_err_engl_es['description_es'].str.replace(i,' ')
    df_err_engl_es['description_en'] = df_err_engl_es['description_en'].str.replace(i,' ')


df_err_engl_es.head(20)
```

```python
df_err_engl_es.to_csv('C:/Users/beder/Desktop/Data 2022/Projects/CIVITATIS-IMPLEMENTATION2022/Outputs/Civitatis Exctraction/civitatis_descrptn_err_solved_2.csv')
```
